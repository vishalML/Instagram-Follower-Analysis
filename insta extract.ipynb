```python
import os
import json
import pandas as pd
```

```python
Path = "D:/Alabs/InstaProject/instaloader/"
```

```python
#initialising

df = pd.DataFrame()
df = df.reset_index(drop=True)

name = []
username = []
private = []
business = []
joined_recently = []

following = []
followers = []
mutual = []
likes = []
comments = []
captions = []
url = []
```

```python
#using os.walk() for iterating through each folder
for root, dirs, files in os.walk(Path):
    for filename in files:
        n = os.path.join(root,filename) #adding root directory and filename to get full path
        if (n.endswith("UTC.json")): #only looking for files ending with "UTC.json" as they contain all the post info
            # Read file
            with open(n) as json_data:
                data = json.load(json_data)
                
                x = data["node"]
                y = x["owner"]
                name.append(y["full_name"])
                username.append(y["username"])
                private.append(y["is_private"])
                business.append(y["is_business_account"])
                joined_recently.append(y["is_joined_recently"])
                following.append(y["edge_follow"]["count"])
                followers.append(y["edge_followed_by"]["count"])
                mutual.append(y["edge_mutual_followed_by"]["count"])
                url.append(x["display_url"])
                likes.append(x["edge_media_preview_like"]['count'])
                comments.append(x["edge_media_to_comment"]["count"])
                
```

```python
df["mutual_connecting"] = mutual
df["total_following"] = following
df["total_followers"] = followers
df["media_url"] = url
df["media_likes"] = likes
df["media_comments"] = comments
df["name"] = name
df["username"] = username
df["private_account"] = private
df["business_account"] = business
df["recently_joined"] = joined_recently
```
