```python
#importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas_profiling as ppf
import seaborn as sns
```

```python
#reading the data
insta = pd.read_csv("./extract_data.csv")
insta.head()
```

```python
insta.columns
```

```python
insta.shape
```

```python
#dropping unnamed column and media url
del insta["Unnamed: 0"]
del insta["media_url"]
```

```python
insta.head()
```

```python
#rearranging columns
insta = insta.iloc[:,[6,0,1,2,3,4,5,7,8,9]]
insta.head()
```

```python
#calculating average likes and comments of each profile
total_likes = []
total_comments = []
total_posts = []
for name in insta.username.unique():
    total_likes.append((insta.media_likes[insta.username==name]).sum())
    total_comments.append((insta.media_comments[insta.username==name]).sum())
    total_posts.append((insta.media_likes[insta.username==name]).count())
```

```python
#only keeping one row per username for better understanding
insta = insta[insta.username.duplicated()==False].reset_index(drop=True)
```

```python
insta.head()
```

```python
#adding total likes, total comments and total posts
insta["total_likes"] = total_likes
insta["total_comments"] = total_comments
insta["total_posts"] = total_posts

#deleting media likes and media comments
del insta["media_likes"]
del insta["media_comments"]
```

```python
insta.head()
```

```python
insta.shape
```

```python
#checking for missing values
insta.isnull().sum()
```

```python
ppf.ProfileReport(insta)
```

```python
#changing true and false to 1 and 0
dict1= { True:1, False:0}
```

```python
insta.columns[(insta.dtypes)==bool]
```

```python
insta["private_account"] = insta["private_account"].map(dict1)
insta["business_account"] = insta["business_account"].map(dict1)
insta["recently_joined"] = insta["recently_joined"].map(dict1)
```

```python
insta.head()
```

Adding followback list to dataframe

```python
follow_back = pd.read_csv("./FollowBack.csv")
follow_back.head()
```

```python
following = list(insta.username)
followers = list(follow_back.username)
```

```python
insta["follow_back"] = [1 if user in followers else 0 for user in following]
insta.head()
```

```python
#converting the clean data to csv
insta.to_csv("./clean_data.csv",index=False)
```
